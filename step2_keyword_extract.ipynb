## 형태소 분석

```python
# import modules
import re
from konlpy.tag import Okt
from collections import Counter
```

```python
# 정규식을 이용한 이모티콘 및 기호 제거
def apply_regular_expression(text):
    hangul = re.compile('[^ ㄱ-ㅣ 가-힣]')  # 한글 추출 규칙: 띄어 쓰기(1 개)를 포함한 한글
    result = hangul.sub('', text)  # 위에 설정한 "hangul"규칙을 "text"에 적용(.sub)시킴
    return result
```

```python
# 한 글자 명사 제거
def available_counter(counter):
    return Counter({x:counter[x] for x in counter if len(x)>1})
available_counter(counter).most_common(100)
```

### 불용어 제거

```python
stopwords = pd.read_csv("https://raw.githubusercontent.com/yoonkt200/FastCampusDataset/master/korean_stopwords.txt").values.tolist()
```

```python
# 데이터셋 특화 불용어 추가
review_stopwords = pd.read_csv("stopword.txt", header=None)
review_stopwords = review_stopwords.iloc[:, 0].tolist()
for word in review_stopwords:
    stopwords.append(word)
```

### Word Count

```python
# BoW 벡터 생성
from sklearn.feature_extraction.text import CountVectorizer

def text_cleaning(review):
    hangul = re.compile('[^ ㄱ-ㅣ 가-힣]') # 정규 표현식 처리
    result = hangul.sub('', review)
    morphs = rhinoMorph.onlyMorph_list(rn, result, pos=['NNG', 'NNP', 'NP', 'VV', 'VA', 'XR', 'IC', 'MM', 'MAG', 'MAJ'], eomi=True)
    morphs = [x for x in morphs if len(x)>1] # 한 글자 키워드 제거
    morphs = [x for x in morphs if x not in stopwords] # 불용어 제거
    return morphs

vect = CountVectorizer(tokenizer = lambda x : text_cleaning(x))
bow_vect = vect.fit_transform(df["review"].tolist())
word_list = vect.get_feature_names_out()
count_list = bow_vect.toarray().sum(axis=0)
```

```python
# 단어 - 빈도 matching
word_count_dict = dict(zip(word_list, count_list))
word_count_dict
```

### TF-IDF

```python
# import modules
from sklearn.feature_extraction.text import TfidfTransformer

tfidf_vectorizer = TfidfTransformer()
tf_idf_vect = tfidf_vectorizer.fit_transform(bow_vect)
```

```python
# 벡터 - 단어 mapping
vect.vocabulary_
```

```python
print(tf_idf_vect)
```

```python
# tf-idf_vect에서 tf-idf 수치값만 추출하여 리스트로 저장
values = tf_idf_vect.data.tolist()

print(values)
```

```python
import numpy as np

# tf-idf_vect는 희소 행렬(Sparse Matrix)이라고 가정합니다.
# (행 인덱스, 열 인덱스) 정보를 추출하여 리스트로 저장
row_indices, column_indices = tf_idf_vect.nonzero()

# tf-idf 수치를 가져와서 리스트로 저장
tf_idf_values = tf_idf_vect.data.tolist()

# tf-idf 수치를 내림차순으로 정렬한 인덱스를 가져옵니다.
sorted_indices = np.argsort(tf_idf_values)[::-1]

# 가장 높은 tf-idf 값의 (행 인덱스, 열 인덱스) 정보를 얻습니다.
highest_tf_idf_index = (row_indices[sorted_indices[0]], column_indices[sorted_indices[0]])

print(f"Highest TF-IDF Value: {tf_idf_values[sorted_indices[0]]}")
print(f"Indices of Highest TF-IDF Value: {highest_tf_idf_index}")
```

### tf-idf 수치 내림차순, 키 맵핑

```python
# 상위n 값을 설정
n = 2000 

# tf-idf_vect는 희소 행렬(Sparse Matrix)이라고 가정
# (행 인덱스, 열 인덱스) 정보를 추출하여 리스트로 저장
row_indices, column_indices = tf_idf_vect.nonzero()

# tf-idf 수치를 가져와서 리스트로 저장
tf_idf_values = tf_idf_vect.data.tolist()

# tf-idf 수치를 내림차순으로 정렬한 인덱스 가져오기
sorted_indices = np.argsort(tf_idf_values)[::-1]

# 상위 n개의 (행 인덱스, 열 인덱스) 정보와 해당 tf-idf 값을 추출
top_n_indices = [(row_indices[idx], column_indices[idx]) for idx in sorted_indices[:n]]
top_n_values = [tf_idf_values[idx] for idx in sorted_indices[:n]]

# 상위 n개의 정보를 출력
for i, (row, col) in enumerate(top_n_indices):
    print(f"Top {i+1}: TF-IDF Value: {top_n_values[i]}, Indices: ({row}, {col})")
```

```python
# vect.vocabulary_에 단어 사전이 저장
# ex. vocabulary = {'apple': 0, 'banana': 1, 'cat': 2, ...}

# 상위 n개의 (행 인덱스, 열 인덱스) 정보가 있다고 가정
# ex. top_n_indices = [(0, 4664), (0, 1777), ...]

# 단어 사전을 이용하여 (행 인덱스, 열 인덱스)에 해당하는 단어 맵핑하기
for row, col in top_n_indices:
    word = list(vect.vocabulary_.keys())[list(vect.vocabulary_.values()).index(col)]
    print(f"({row}, {col}): {word}")
```

```python
# vect.vocabulary_에 단어 사전이 저장되어 있다고 가정합니다.
# 예시: vocabulary = {'apple': 0, 'banana': 1, 'cat': 2, ...}

# 상위 n개의 (행 인덱스, 열 인덱스) 정보가 있다고 가정합니다.
# 예시: top_n_indices = [(0, 4664), (0, 1777), ...]

# 원하는 순위의 범위를 설정합니다.
start_rank = 1000
end_rank = 2000

# 순위 중간에 해당하는 (행 인덱스, 열 인덱스) 정보를 추출합니다.
middle_rank_indices = top_n_indices[start_rank-1:end_rank]

# 단어 사전을 이용하여 (행 인덱스, 열 인덱스)에 해당하는 단어를 알아냅니다.
for i, (row, col) in enumerate(middle_rank_indices, start=start_rank):
    word = list(vect.vocabulary_.keys())[list(vect.vocabulary_.values()).index(col)]
    print(f"Rank {i}: ({row}, {col}): {word}")
```
