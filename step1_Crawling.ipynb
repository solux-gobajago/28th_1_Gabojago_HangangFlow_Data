```python
def crawling(link, n):
    # dateframe ìƒì„±
    data = pd.DataFrame(data=[], columns=['review','rate'])
    
    # connect driver
    service = Service('C:/Users/chromedriver.exe')
    driver = webdriver.Chrome(service=service)
    driver.get(link)
    time.sleep(4)

    review_list=[]
    # ìŠ¤í¬ë¡¤
    for i in range(n): # ë¦¬ë·° ê°¯ìˆ˜ì— ë”°ë¼ range ë²”ìœ„ ìˆ˜ì •
        try:
            scroll = driver.find_element(By.CLASS_NAME, 'm6QErb.DxyBCb.kA9KIf.dS8AEf')
            time.sleep(1)
            driver.execute_script('arguments[0].scrollTop = arguments[0].scrollHeight', scroll)
            location1 = scroll.location_once_scrolled_into_view
            print(i, location1)
        except Exception as e:
            print(e)
            break

        time.sleep(1)

    i = 0 # ìŠ¤í¬ë¡¤ ë‹¤ìš´ íšŸìˆ˜ë¥¼ ì €ì¥í•˜ê¸° ìœ„í•œ ë³€ìˆ˜
    while True:
        try:
            # ë”ë³´ê¸° ë²„íŠ¼ì„ ì°¾ìŠµë‹ˆë‹¤.
            more_btn = driver.find_element(By.CLASS_NAME, "w8nwRe.kyuRq")
            driver.execute_script("arguments[0].click();", more_btn) # ë”ë³´ê¸° ë²„íŠ¼ì„ í´ë¦­
            i += 1 # ìŠ¤í¬ë¡¤ ë‹¤ìš´ íšŸìˆ˜ë¥¼ 1 ì¦ê°€ì‹œí‚µë‹ˆë‹¤.
            print(f"{i}ë²ˆì§¸ ìŠ¤í¬ë¡¤ ë‹¤ìš´ ì™„ë£Œ")
            time.sleep(5) # ìƒˆë¡œìš´ ì •ë³´ê°€ ë¡œë”©ë  ìˆ˜ ìˆë„ë¡ 5ì´ˆì˜ ë”œë ˆì´
        except NoSuchElementException: # ë” ì´ìƒ ë”ë³´ê¸° ë²„íŠ¼ì´ ì—†ë‹¤ë©´ ì¢…ë£Œ
            print("ë” ì´ìƒ ë”ë³´ê¸° ë²„íŠ¼ì´ ì—†ìŠµë‹ˆë‹¤.")
            break

    review=driver.find_elements(By.CLASS_NAME, 'wiI7pd')
    rate=driver.find_elements(By.CLASS_NAME, 'kvMYJc')
#     print(len(review))
#     print(len(rate))
    for i in range(len(review)):
        tmp = []
        tmp.append(review[i].text)
        tmp.append(rate[i].get_attribute('aria-label'))

        tmp = pd.DataFrame(data=[tmp], columns = data.columns)
        data = pd.concat([data, tmp])
    
    # ì¸ë±ìŠ¤ ì´ˆê¸°í™”
    data.reset_index(inplace=True, drop=True)
    
    # disconnect driver
    driver.quit()
    
    return data
```
